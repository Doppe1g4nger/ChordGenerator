```python
import os
import re

import music21

harmony_fp = input("Input directory of musicXML harmony files: ")

melody_fp = input("Input file or directory of musicXML melody files: ")
    
if os.path.isdir(melody_fp):
    melody_list = [os.path.join(melody_fp, path) for path in os.listdir(melody_fp)]
else:
    melody_list = [melody_fp]
```

```python
save_dir = input("Input the directory to save the generated musicxml files: ")
if not os.path.exists(save_dir):
    os.mkdir(save_dir)
```

```python
temperley_harmony = True

use_declerq_harmony = input("Would you like to use the deClerq harmony files (Temperley used by default) (y/n): ")

if use_declerq_harmony == "y":
    temperley_harmony = False
```

```python
if temperley_harmony:
    file_suffix = "_dt.musicxml"
else:
    file_suffix = "_tdc.musicxml"
```

```python
for melody in melody_list:
    song_name = melody.split("\\")[-1].split(".")[0]
    print("Processing:", song_name)
    
    if song_name.endswith("_dt"):
        song_name = song_name[:-3]
    elif song_name.endswith("_tdc"):
        song_name = song_name[:-4]
        
    melody_score = music21.converter.parse(melody)
    harmony_score = music21.converter.parse(os.path.join(harmony_fp, song_name + file_suffix))
    
    num_melody_measures = len(melody_score.parts[0].measures(0, None))
    num_harmony_measures = len(harmony_score.parts[0].measures(0, None))
    
    if num_harmony_measures > num_melody_measures:
        rest_measure = music21.stream.Measure()
        time_sig = melody_score.parts[0].measures(0, None)[-1].timeSignature
        
        if time_sig is not None:
            rest_measure.timeSignature = time_sig
        else:
            rest_measure.timeSignature = music21.meter.TimeSignature("4/4")
            
        rest_measure.append(music21.note.Rest(quarterLength=(
            4 * rest_measure.timeSignature.numerator / rest_measure.timeSignature.denominator
        )))
        melody_score.parts[0].repeatAppend(rest_measure, num_harmony_measures - num_melody_measures)
                                
    elif num_melody_measures > num_harmony_measures:
        rest_measure = music21.stream.Measure()
        time_sig = harmony_score.parts[0].measures(0, None)[-1].timeSignature
        
        if time_sig is not None:
            rest_measure.timeSignature = time_sig
        else:
            rest_measure.timeSignature = music21.meter.TimeSignature("4/4")
            
        rest_measure.append(music21.note.Rest(quarterLength=(
            4 * rest_measure.timeSignature.numerator / rest_measure.timeSignature.time_sig.denominator
        )))
        harmony_score.parts[0].repeatAppend(rest_measure, num_melody_measures - num_harmony_measures)
        
    
    melody_score.insert(0, harmony_score.parts[0])
    
    GEX = music21.musicxml.m21ToXml.GeneralObjectExporter(melody_score)
    
    out = GEX.parse()
    musicxml = out.decode('utf-8').strip()
    
    with open(os.path.join(save_dir, song_name + ".musicxml"), "w") as outfile:
        outfile.write(musicxml)
```
